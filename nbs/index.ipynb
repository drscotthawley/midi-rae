## Usage

### Installation

Install latest from the GitHub [repository][repo]:

```sh
$ pip install git+https://github.com/drscotthawley/midi-rae.git
```

or from [pypi][pypi]:

```sh
$ pip install midi-rae
```

### GPU Acceleration (Optional)

For faster UMAP and PCA projections using NVIDIA GPUs, install with the `gpu` extra:

```sh
$ pip install midi-rae[gpu]
```

This requires [RAPIDS](https://rapids.ai/) cuML and cuPy. See the [RAPIDS installation guide](https://docs.rapids.ai/install) for setup instructions — pip wheels are available for supported CUDA versions:

```sh
$ pip install cuml-cu12 cupy-cuda12x  # for CUDA 12
```

Without GPU dependencies, the package falls back to CPU-based sklearn and umap-learn automatically.
